```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(20)
```

```python
cluster_1 = np.random.normal(loc=[2, 2], scale=1.0, size=(100, 2))
cluster_2 = np.random.normal(loc=[7, 5], scale=1.0, size=(100, 2))
cluster_3 = np.random.normal(loc=[12, 3], scale=1.0, size=(100, 2))
```

```python
x = np.vstack((cluster_1, cluster_2, cluster_3))
```

```python
plt.scatter(cluster_1[:, 0], cluster_1[:, 1], color="blue", label="Cluster 1")
plt.scatter(cluster_2[:, 0], cluster_2[:, 1], color="green", label="Cluster 2")
plt.scatter(cluster_3[:, 0], cluster_3[:, 1], color="red", label="Cluster 3")
plt.title("Generated Clusters")
plt.legend()
plt.show()
```

```python
plt.scatter(x[:, 0], x[:, 1], color="black")
plt.title("Clusters")
plt.legend()
plt.show()
```

```python
y_1 = np.ones(shape=(100, 1))
y_2 = np.full(shape=(100, 1), fill_value=2)
y_3 = np.full(shape=(100, 1), fill_value=3)
y = np.concatenate((y_1, y_2, y_3), axis = 0)
```

```python
arr = np.concatenate((x, y), axis = 1)
df = pd.DataFrame(arr, columns=["f1", "f2", "target"])
df.sample(10)
```

```python
df.to_csv("../data/dataset.csv", index=False)
```
